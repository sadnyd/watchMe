```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import requests
import os
import csv
import time

# List of brands for downloading images
BRANDS = [
    'rolex', 'audemarspiguet', 'breitling', 'iwc', 'jaegerlecoultre',
    'omega', 'panerai', 'patekphilippe', 'cartier', 'gucci',
    'seiko', 'movado', 'zenith'
]

# Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"
chrome_options.add_argument(f"user-agent={user_agent}")

# Specify the path to the ChromeDriver
driver = webdriver.Chrome(options=chrome_options)

# Ensure the images directory exists
os.makedirs("images", exist_ok=True)

for brand in BRANDS:
    try:
        # Open the corresponding .txt file for each brand
        with open(f"/content/data/{brand}.txt", newline='') as csvfile:
            data = csv.reader(csvfile)

            for index, item in enumerate(data):
                image_url = item[0]
                image_name = item[1]

                try:
                    # Use Selenium to open the image URL
                    driver.get(image_url)

                    # Find the image element (update selector based on website structure)
                    image_element = driver.find_element("tag name", "img")
                    actual_image_url = image_element.get_attribute("src")

                    # Add headers to the request
                    headers = {
                        "User-Agent": user_agent,
                        "Referer": image_url  # Set referer as the page URL
                    }
                    # Download the image using requests with headers
                    response = requests.get(actual_image_url, headers=headers, stream=True)
                    if response.status_code == 200:
                        with open(f"images/{brand}-{index+1}-{image_name}.jpg", "wb") as file:
                            file.write(response.content)

                        print(f"Downloaded {brand}-{index+1}-{image_name}.jpg")
                    else:
                        print(f"Failed to download {brand}-{index+1}-{image_name}.jpg, status code: {response.status_code}")

                        # Optional fallback: Save a screenshot if direct download fails
                        driver.save_screenshot(f"images/{brand}-{index+1}-{image_name}_screenshot.png")
                        print(f"Saved screenshot as fallback for {brand}-{index+1}-{image_name}.jpg")

                except Exception as e:
                    print(f"Error downloading {image_url}: {e}")

    except FileNotFoundError:
        print(f"File data/{brand}.txt not found. Please upload it to the correct directory.")

# Close the browser session after downloading all images
driver.quit()
```
