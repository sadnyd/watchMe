```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os
import csv
import time

# List of brands for downloading images
BRANDS = [
    'rolex', 'audemarspiguet', 'breitling', 'iwc', 'jaegerlecoultre',
    'omega', 'panerai', 'patekphilippe', 'cartier', 'gucci',
    'seiko', 'movado', 'zenith'
]

# Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36")

# Specify the path to the ChromeDriver
driver = webdriver.Chrome(options=chrome_options)


# Ensure the images directory exists
os.makedirs("images", exist_ok=True)

for brand in BRANDS:
    try:
        # Open the corresponding .txt file for each brand
        with open(f"/content/data/{brand}.txt", newline='') as csvfile:
            data = csv.reader(csvfile)

            for index, item in enumerate(data):
                image_url = item[0]
                image_name = item[1]

                try:
                    # Use Selenium to open the image URL
                    driver.get(image_url)

                    # Get the image content
                    with open(f"images/{brand}-{index+1}-{image_name}.jpg", "wb") as file:
                        file.write(driver.page_source.encode('utf-8'))

                    print(f"Downloaded {brand}-{index+1}-{image_name}.jpg")
                    #time.sleep(1)  # Slow down requests

                except Exception as e:
                    print(f"Error downloading {image_url}: {e}")

    except FileNotFoundError:
        print(f"File data/{brand}.txt not found. Please upload it to the correct directory.")

# Close the browser session after downloading all images
driver.quit()
```
